# Using Histogram, determine the distribution of the numerical columns in the dataset.

```python
numerical_columns = heart.select_dtypes(include=['int64', 'float64'])

num_cols = numerical_columns.columns
num_cols_count = len(num_cols)
num_cols_per_row = 3  
num_rows = (num_cols_count - 1) // num_cols_per_row + 1

fig, axes = plt.subplots(num_rows, num_cols_per_row, figsize=(20, 15))
fig.subplots_adjust(hspace=1.0)

for i, col in enumerate(num_cols):
    row = i // num_cols_per_row
    col_num = i % num_cols_per_row
    ax = axes[row, col_num]

   
    ax.hist(heart[col], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')


for i in range(num_cols_count, num_rows * num_cols_per_row):
    fig.delaxes(axes.flatten()[i])

plt.show()
```

# Check using proper visualizations, whether serum cholesterol level affects the blood pressure 

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Serum Cholesterol\n (in mg/dl)', y='Rest \nBlood Pressure', data=heart, alpha=0.6, hue='Sex')
plt.title('Scatter Plot: Serum Cholesterol vs. Blood Pressure')
plt.xlabel('Serum Cholesterol (mg/dl)')
plt.ylabel('Blood Pressure')


plt.figure(figsize=(10, 6))
sns.boxplot(x='Serum Cholesterol\n (in mg/dl)', y='Rest \nBlood Pressure', data=heart)
plt.title('Box Plot: Serum Cholesterol vs. Blood Pressure')
plt.xlabel('Serum Cholesterol (mg/dl)')
plt.ylabel('Blood Pressure')

plt.show()
```
